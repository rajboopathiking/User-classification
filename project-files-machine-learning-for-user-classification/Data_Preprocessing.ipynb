```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split

class DataPreparing:
    def __init__(self, file_path):
        self.df = pd.read_csv(file_path)

    def outlier_removal(self, col):
        q1 = self.df[col].quantile(0.25)
        q3 = self.df[col].quantile(0.75)
        iqr = q3 - q1
        low = q1 - 1.5 * iqr
        high = q3 + 1.5 * iqr
        self.df = self.df[(self.df[col] > low) & (self.df[col] < high)]

    def data_cleaning(self, col, var_type="categorical"):
        if var_type == "categorical":
            self.df[col].fillna(self.df[col].mode()[0], inplace=True)
        else:
            self.df[col].fillna(self.df[col].mean(), inplace=True)

    def prepare_data(self):
        if self.df.duplicated().sum() > 0:
            self.df.drop_duplicates(inplace=True)

        label = OrdinalEncoder()
        self.df["student_country_enc"] = label.fit_transform(
            self.df["student_country"].values.reshape(-1, 1)
        )

        X = self.df.drop(["purchased", "student_country", "student_country_enc"], axis=1)
        y = self.df["purchased"]

        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=0.2, stratify=y
        )
        return X_train, X_test, y_train, y_test
```

```python
data_handler = DataPreparing("/Users/godfather_101/Boopathi_workspace/Projects/365datascience/project-files-machine-learning-for-user-classification/ml_datasource.csv")
X_train, X_test, y_train, y_test = data_handler.prepare_data()
```
